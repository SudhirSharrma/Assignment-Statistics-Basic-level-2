```python
#converted above data into .csv removed percentage sympbol
dataq1=pd.read_csv("q1_level2.csv")
dataq1.head()
```

```python
#creating box plot for above data
sn.boxplot(x=dataq1['Measure X'])
```

```python
#using describe function to get basic details on data
dataq1.describe()
```

```python
#calculating Variance(σ^2)
dataq1['Measure X'].var()
```

```python
#identifying 1st quantile and 3rd quantile
print("1st Quantile: ",dataq1.quantile(0.25))
print("3rd Quantile: ",dataq1.quantile(0.75))
```

```python
#identifying Interquantile range(IQR)
Q1=dataq1.quantile(0.25)
Q3=dataq1.quantile(0.75)
IQR=Q3-Q1
print ("Interquantile Range (IQR) is: ",IQR)
```

```python
#finding lower limt and upper lmit
low_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
```

```python
low_limit
```

```python
upper_limit
```

```python
#identifyhing outliers
outlier_low=(dataq1['Measure X']<float(low_limit))
outlier_upper=(dataq1['Measure X']>float(upper_lim))
```

```python
#extracting outliers from data
dataq1['Measure X'][outlier_low|outlier_upper]
```
